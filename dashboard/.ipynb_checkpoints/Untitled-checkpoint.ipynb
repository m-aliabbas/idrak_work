```python
checkpoints_paths=glob.glob('checkpoints/*.ckpt')
```

```python
trained_models=glob.glob('models/*/')
```

```python
trained_models_names=[i[7:].replace('/','') for i in trained_models]
```

```python
def list_contain(text,base_array):
    temp_list=[i for i in base_array if text in i ]
    return temp_list
```

```python
selected_ckpt=list_contain(trained_models_names[0],checkpoints_paths)
```

```python
selected_ckpt_dict={os.path.basename(i)[:-5]:i for i in selected_ckpt}
```
