```python
recording_dir = "recordings/"
```

```python
def downloadLink(URL):
    try:
        headers={'User-Agent':'test-app/1.0'}  #This header will prevent invalid response      
        r = requests.get(URL, timeout=10, headers=headers, stream = True)
        filename = URL.split("/")[-1]
        if r.status_code == 200:
        
            # creating the output file in write binary mode, because the data we get from internet is in binary
            with open(recording_dir +filename, 'wb') as f:
                for chunk in r.iter_content(chunk_size=1024):
                    # writing one chunk at a time to mp3 file
                    if chunk:
                        f.write(chunk)
            print(True)
            return True
    except Exception as e:
        print('Error',e)
        return False            
```
